### Question No : 02 : Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
# Solution No :- 02  ( By Using , For_Loop statement & Python commands)

#Importing the Required Libraries.
from bs4 import BeautifulSoup
import requests

#Third Step - Sending Get -request to the webpage server to fetch SourceCode of the page.
page3 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
page4 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")

#Calling the get_request variable "page3 & page4 " , to fetch SourceCode pages permissions.
page3
page4

soup3 = BeautifulSoup(page3.content)    # Storing the Source Code Contents , to variable 'soup3'
soup4 = BeautifulSoup(page4.content)    # Storing the Source Code Contents , to variable 'soup4'

soup3                 # Calling the variable 'soup3'
soup4                 # Calling the variable 'soup4'


moviesnames = []
for i in soup3.find_all("h3",class_="lister-item-header"):
    moviesnames.append(i.text)
for i in soup4.find_all("h3",class_="lister-item-header"):
    moviesnames.append(i.text)

    
year = []
for i in soup3.find_all("span",class_="lister-item-year text-muted unbold"):
    year.append(i.text)
for i in soup4.find_all("span",class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
ratings = []
for i in soup3.find_all("div",class_="inline-block ratings-imdb-rating"):
    ratings.append(i.text)
for i in soup4.find_all("div",class_="inline-block ratings-imdb-rating"):
    ratings.append(i.text)


import pandas as pd
df = pd.DataFrame({'MOVIE_NAME':moviesnames,'YEAR_OF_RELEASE':year,'RATE_SCORE':ratings})


print("To display IMDB’s Top rated 100 movies are as below's : ",sep="\n\n")
df
```

### Question No : 03 : Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
# Solution No :- 03  ( By Using , For_Loop statement & Python commands)

#Importing the Required Libraries.
from bs4 import BeautifulSoup
import requests

#Third Step - Sending Get -request to the webpage server to fetch SourceCode of the page.
page = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")

#Calling the get_request variable "page " , to fetch SourceCode pages permissions.
page

soup = BeautifulSoup(page.content)    # Storing the Source Code Contents , to variable 'soup'

soup                 # Calling the variable 'soup'

moviesnames = []

for i in soup.find_all("td",class_="titleColumn"):
    moviesnames.append(i.text)
    
moviesnames


    
year = []

for i in soup.find_all("span",class_="secondaryInfo"):
    year.append(i.text)
    
    
year
    
    
ratings = []

for i in soup.find_all("td",class_="ratingColumn imdbRating"):
    ratings.append(i.text)

ratings

import pandas as pd
df = pd.DataFrame({'INDIAN_MOVIE_NAMES':moviesnames,'YEAR_OF_RELEASE':year,'RATE_SCORES':ratings})

print("To display IMDB’s Top rated 100 INDIAN MOVIES are as below's : ",sep="\n\n")
df
```
