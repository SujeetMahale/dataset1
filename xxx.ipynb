```python
# Solution No :- 02  ( By Using , For_Loop statement & Python commands)

#Importing the Required Libraries.
from bs4 import BeautifulSoup
import requests

#Third Step - Sending Get -request to the webpage server to fetch SourceCode of the page.
page3 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")

#Calling the get_request variable "page" , to fetch SourceCode pages permissions.
page3       

soup3 = BeautifulSoup(page3.content)    # Storing the Source Code Contents , to variable 'soup'

soup3                 # Calling the variable 'soup'

moviesnames = []

for i in soup3.find_all("h3",class_="lister-item-header"):
    moviesnames.append(i.text)   

year = []

for i in soup3.find_all("span",class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
ratings = []

for i in soup3.find_all("div",class_="inline-block ratings-imdb-rating"):
    ratings.append(i.text)


import pandas as pd
df = pd.DataFrame({'MOVIE':moviesnames,'YEAR':year,'RATE':ratings})

df
```
